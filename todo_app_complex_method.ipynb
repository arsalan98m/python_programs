```python
# Todo App in python 
# Main Menu
# Add Tasks
# View Tasks
# Delete Tasks
# Terminate App

# check user input
def isInteger(userInput):
    try:
        int(userInput)
        return True
    except ValueError:
        return False
    
    
def mainMenu(menu):
    print("""
---------------
|  Main Menu  |
---------------
    """)
    for menu_key,menu_name in menu.items():
        print(f"{menu_key}. {menu_name}")
    userCommand = input("Enter Command: ")
    return userCommand    

def addTask(taskDictionary):
    print("""
---------------
| Adding Task |
---------------""")
    taskIndex = len(taskDictionary) + 1
    taskName = input(f"Task #{taskIndex}: ")
    taskDictionary.update({taskIndex: taskName.title()})
    print(f"Task #{taskIndex} Successfully added. ")
    
def viewTask(taskDictionary):
    print("""
----------------
| Viewing Tasks |
----------------
    """)
    if len(taskDictionary) > 0:
        for taskIndex, taskName in taskDictionary.items():
            print(f"Task #{taskIndex} {taskName}")
    else:
        print("No Tasks Found")
    
def deleteTask(taskDictionary):
    print("""
----------------
| Deleting Tasks |
----------------
    """)
    if len(taskDictionary) > 0:
        for taskIndex, taskName in taskDictionary.items():
            print(f"Task #{taskIndex} - {taskName}")
        
        taskIndex = int(input("Enter Task Index eg.1, 2: "))
        if taskIndex in taskDictionary.keys():
            taskDictionary.pop(taskIndex)
            print(f"Task #{taskIndex} - Successfully Deleted.")
        else:
            print("Invalid Index, Try valid Index.")
    else:
        print("No Tasks found.")

def deleteAllTasks(taskDictionary):
    if len(taskDictionary) > 0:
        taskDictionary.clear()
        print("Deleted All Task Sucessfully.")
        
menu = {
    1: "Add Tasks",
    2: "View Tasks",
    3: "Delete Tasks",
    4: "Delete All Tasks",
    5: "Exit",
    
}



taskDictionary = {}

while True:
    userCommand = mainMenu(menu)
    
    if isInteger(userCommand) == True:
        userCommand = int(userCommand)
        if userCommand > 0 and userCommand <= 5:
            # Add Task
            if userCommand == 1:
                addTask(taskDictionary)
            
            # View Task
            if userCommand == 2:
                viewTask(taskDictionary)
            
            # Delete Task
            if userCommand == 3:
                deleteTask(taskDictionary)
            
            # Delete All task
            if userCommand == 4:
                deleteAllTasks(taskDictionary)
            # Exit 
            if userCommand == 5:
                print("Thank you!")
                break
        else:
            print("Invalid Command, Try valid Command")
    else:
        print("Invalid Command, try")
```
