```python
questions = [
    {
        "Qno": 1,
        "question": "What is python?",
        "options": ["Interpreted language", "Compiled language", "Low-level language", "Assembly language"],
        "answer" : "1",
    },
    {
        "Qno": 2,
        "question": "Which one is NOT a legal variable name?",
        "options": ["my-var", "_myvar", "MYvar", "my_var"],
        "answer": "1",
    },
    {
        "Qno": 3,
        "question": "How do you create a variable with the numeric value 5?",
        "options": ["x = 5", "x = int(5)", "x = int('5')", "Both Correct"],
        "answer": "4",
    },
    {
        "Qno": 4,
        "question": "What is the correct file extension for python files?",
        "options": [".pt", ".pyth", ".py", ".pyt"],
        "answer": "3",
    },
    {
        "Qno": 5,
        "question": "What method can be used to remove any white space from both the beginnig and end of a string?",
        "options": ["ptrim()","strip()", "len()", "trim"],
        "answer": "2",
    }
]


```

```python
score = 0

for question in questions:
    output_data = "\n\n"+str(question["Qno"]) + ": "+ question["question"] +"\n\n"+"Choose correct answer: "+"\n"+"1: "+question["options"][0] +"\n"+ "2: "+question["options"][1] + "\n" + "3: " +question["options"][2]+"\n"+"4: "+question["options"][3]
    answer = input(output_data)
    
    if answer == question['answer']:
        score +=1

# Score Condition
if score == 5:
    print("\n\nYour Score is: "+str(score))
    print("üéâüéâüçï YOU WIN!!!! üéâüéâüçï")
elif score == 3:
    print("\n\nYour Score is: "+str(score))
    print("üòéCool Keep learning")
else:
    print("\n\nYour Score is: "+str(score))
    print("Keep Practicing")

# Calculation marks
#total marks / obt * 100
percentage = score/5 * 100
print("Your Percentage is : "+str(percentage)+"%")
```
